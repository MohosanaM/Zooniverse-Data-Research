```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# 1. Directory Correction: Ensure we are inside the repository folder
if os.path.exists('Zooniverse-Data-Research'):
    if not os.getcwd().endswith('Zooniverse-Data-Research'):
        os.chdir('Zooniverse-Data-Research')

# 2. File Path Definition (Matching your actual filename in the 'data' folder)
file_path = 'data/classification_audit_100.csv'

# Debugging: Confirming what Colab sees
print(f"✅ Current working directory: {os.getcwd()}")
if os.path.exists(file_path):
    print(f"✅ File found: {file_path}")
else:
    print(f"❌ File NOT found at: {file_path}")
    print(f"Available files in data/: {os.listdir('data') if os.path.exists('data') else 'Data folder missing'}")

# 3. Processing and Visualization
try:
    # Load the data
    df = pd.read_csv(file_path)

    # We use 'Galaxy_Type' as that was the column name in our generated CSV
    if 'Galaxy_Type' in df.columns:
        type_counts = df['Galaxy_Type'].value_counts()

        # Professional "Deep Space" Visualization
        plt.style.use('dark_background')
        plt.figure(figsize=(10, 6))

        # Using a sleek color palette
        colors = ['#00ffcc', '#fe53bb', '#08f7fe', '#f5d300', '#00ff41']
        type_counts.plot(kind='bar', color=colors, edgecolor='white')

        plt.title('Final Audit: Morphological Distribution (N=100)', fontsize=14, fontweight='bold', color='white')
        plt.xlabel('Galaxy Morphology', fontsize=12, color='white')
        plt.ylabel('Frequency', fontsize=12, color='white')
        plt.xticks(rotation=45)
        plt.grid(axis='y', linestyle='--', alpha=0.3)

        # Save to the visualizations folder
        os.makedirs('visualizations', exist_ok=True)
        plt.tight_layout()
        plt.savefig('visualizations/galaxy_distribution_100.png')
        plt.show()

        print(f"📊 Success! Visualization saved to 'visualizations/galaxy_distribution_100.png'")
    else:
        print(f"❌ Error: Column 'Galaxy_Type' not found. Available columns: {list(df.columns)}")

except Exception as e:
    print(f"❌ An unexpected error occurred: {e}")
```
